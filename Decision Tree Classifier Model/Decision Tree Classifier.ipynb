# Decision Tree Classifier Model

```python
#importing all the required files

%matplotlib inline
import pandas as pd
import numpy as np

data=pd.read_csv('Iris.csv')
data.head()
```

```python
X=data.iloc[:,0:5]
Y=data.iloc[:,-1]
new_data={"Iris-setosa":"0","Iris-versicolor":"1","Iris-virginica":"2"}
data['target']=data['Species'].map(new_data)
y=data.iloc[:,-1]
```

```python
from sklearn.tree import DecisionTreeClassifier


# Defining the model with default hyperparameters
clf = DecisionTreeClassifier(random_state=42)

# Fit the model
clf.fit(X, y)
```

```python
#Visualizing the Decision Tree graphically
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


X_raw=data.iloc[:,1:5]
f_name=X_raw.columns.tolist()
col_name=data['Species'].unique().tolist()
plt.figure(figsize=(25,15))
tree_image = plot_tree(clf,feature_names=f_name,class_names=col_name,max_depth=5,precision=4,label="all",filled=True,rounded=True)
plt.savefig('tree_image')
```
